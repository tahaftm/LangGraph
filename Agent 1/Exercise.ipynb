```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name:str
    message:str

def get_complement(state: AgentState) ->AgentState:
    state['message'] = "Hey " + state['name'] + ", You are good at coding"

    return state
```

```python
graph = StateGraph(state_schema=AgentState)
graph.add_node("complement",get_complement)
graph.set_entry_point('complement')
graph.set_finish_point('complement')
app = graph.compile()
```

```python
app
```

```python
result = app.invoke({'name':'Taha'})
```
